**Summary**

- Images are greyscaled

```python
import pandas as pd
import numpy as np

import cv2

import matplotlib.pyplot as plt
```

```python
# Read the annotations file that contains the label and the image file name
labels = pd.read_csv('../data/raw/annotations.csv', header=None, names=['fname', 'label'])
```

```python
display(labels.head()); print(labels.shape); display(labels.label.value_counts())
```

We see that the data isn't shuffled. We also see that there is a total of 400 labels in this dataset with 200 pictures for each target class.

# Explore what an image is

```python
img = cv2.imread(f"../data/raw/all/{labels['fname'].iloc[0]}")
height, width, channels = img.shape
print(f'height = {height}, width = {width}, channels = {channels}')
print(img[0][0])
```

We see here that an image has the dimensions 193 X 162 pixels. The image has three channels, meaning it is in the RGB format. However, the pixels are all the same in the three channels, which menas that the image is in greyscale.

# View images

```python
figure, axes = plt.subplots(2, 3)

for i in range(3):
    
    # Show neutral images
    img = cv2.imread(f"../data/raw/all/{labels['fname'].iloc[i]}", 0) # reads as grayscale
    axes[0,i].imshow(img, cmap='gray')
    axes[0,i].set_title('neutral')
    axes[0,i].set_axis_off()
    
    # Show smilling images
    img = cv2.imread(f"../data/raw/all/{labels['fname'].iloc[200+i]}", 0) # reads as grayscale
    axes[1,i].imshow(img, cmap='gray')
    axes[1,i].set_title('smiling')
    axes[1,i].set_axis_off()
```
